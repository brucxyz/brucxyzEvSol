3.1. Reformat Dataframe

```python
from datetime import datetime

# Convert the 'day' column to datetime format
df['day'] = pd.to_datetime(df['day'])

# Convert the datetime format to year-month-day format
df['day'] = df['day'].dt.strftime('%Y-%m-%d')
```

```python
df['day'] = pd.to_datetime(df['day'])

# Select all rows with a date year-12-31
df_y = df.loc[(df['day'].dt.month == 12) & (df['day'].dt.day == 31)]

# Print the selected rows
df_y
```

```python
# Drop columns ending with "wind_capacity"

df_y = df_y.drop(df_y.filter(like='wind_capacity').columns, axis=1)
```

```python
df_y.columns
```

```python
df_y
```

```python
# Create Total capacity of ALL for each year

selected_columns = df_y.iloc[:, 1:]

df_y['All_total'] = selected_columns.sum(axis=1)

df_y
```

```python
# Check if All Total correct in random row

row = df_y.iloc[3]
row
```

```python
# Select DK

columns_DK = ['day', 'All_total', 'DK_solar_capacity', 'DK_wind_offshore_capacity', 'DK_wind_onshore_capacity']
DK = df_y.loc[:, columns_DK]

# Drop rows where Solar Capacity is 0
DK_Sol = DK.drop(DK[DK["DK_solar_capacity"] == 0].index)

DK_Sol
```

```python
# Create Total capacity and percentage of Solar capacity of DK

DK_selected_columns = ['DK_solar_capacity', 'DK_wind_offshore_capacity', 'DK_wind_onshore_capacity']

DK_Sol['DK_total'] = DK_Sol[DK_selected_columns].sum(axis=1)

DK_Sol['DK_sol_percentage'] = DK_Sol['DK_solar_capacity'] / DK_Sol['DK_total'] * 100

DK_Sol['DK_sol_percentage_Total'] = DK_Sol['DK_solar_capacity'] / DK_Sol['All_total'] * 100

DK_Sol_Tot = DK_Sol.drop(['DK_wind_offshore_capacity', 'DK_wind_onshore_capacity'], axis=1)

DK_Sol_Tot
```

```python
# Display information about the DataFrame, such as data types and missing values

DK_Sol_Tot.info()
```

```python
# Display Data types of these columns

DK_Sol_Tot.dtypes
```

```python
# Display basic statistics of the numerical columns

DK_Sol_Tot.describe()
```

```python
# Select FR
columns_FR = ['day', 
              'All_total',
              'FR_bioenergy_capacity',
              'FR_geothermal_capacity', 
              'FR_hydro_capacity', 
              'FR_marine_capacity',
              'FR_solar_capacity', 
              'FR_wind_onshore_capacity']
FR = df_y.loc[:, columns_FR]

# Drop rows where Solar Capacity is 0
FR_Sol = FR.drop(FR[FR["FR_solar_capacity"] == 0].index)

FR_Sol
```

```python
# Create Total capacity and percentage of Solar capacity of FR
FR_selected_columns = ['FR_bioenergy_capacity', 
                       'FR_geothermal_capacity', 
                       'FR_hydro_capacity', 
                       'FR_marine_capacity', 
                       'FR_solar_capacity',
                       'FR_wind_onshore_capacity']

FR_Sol['FR_total'] = FR_Sol[FR_selected_columns].sum(axis=1)

FR_Sol['FR_solar_percentage'] = FR_Sol['FR_solar_capacity'] / FR_Sol['FR_total'] * 100

FR_Sol['FR_solar_percentage_Total'] = FR_Sol['FR_solar_capacity'] / FR_Sol['All_total'] * 100

FR_Sol_Tot = FR_Sol.drop(['FR_bioenergy_capacity', 
                          'FR_geothermal_capacity', 
                          'FR_hydro_capacity', 
                          'FR_marine_capacity',
                          'FR_wind_onshore_capacity'], axis=1)

FR_Sol_Tot
```

```python
# Join DK & FR

DKFR = pd.merge(DK_Sol_Tot, FR_Sol_Tot)

DKFR
```

```python
DKFR.columns
```

```python
# Rename Columns

DKFR = DKFR.rename(columns={'day': 'Day', 
                            'All_total': 'Tot_cap', 
                            'DK_solar_capacity': 'DK_sol_cap',
                            'DK_sol_percentage': 'DK_sol%',
                            'DK_sol_percentage_Total': 'DK_sol%_total',
                            'FR_solar_capacity': 'FR_sol_cap',
                            'FR_solar_percentage': 'FR_sol%',
                            'FR_solar_percentage_Total': 'FR_sol%_total'})

DKFR
```

```python
# Describe

DKFR.describe()
```

```python
# outliners

import pandas as pd
from scipy import stats

# Load the time series data
df = DKFR

# Calculate the modified z-score for each data point
df['z_score'] = stats.zscore(df['DK_sol_cap'])

# Define a threshold for outlier detection
threshold = 0

# Identify the outliers based on the threshold
outliers = df[abs(df['z_score']) > threshold]

# Print the outliers
outliers
```

```python
# Plot Chart of DK & FR solar capacity

day = DKFR['Day']
DK_solar_capacity = DKFR['DK_sol_cap']
FR_solar_capacity = DKFR['FR_sol_cap']

plt.figure(figsize=(10, 6))
plt.plot(day, DK_solar_capacity, label='DK_solar_capacity')
plt.plot(day, FR_solar_capacity, label='FR_solar_capacity')

# This Plot Chart shows the difference in between DK and FR.
# FR started major increasment of the Solar Stations installation in 2004, 8 years earlier than DK in 2012, and it has huge Solar capacity compare to DK.
# FR Solar Capacity increasment also last long for 15 years, where the increasment of Solar in DK last for 4 years. 
```

```python
# Plot Chart of DK & FR solar total percentage

day = DKFR['Day']
DK_solar_capacity = DKFR['DK_sol%']
FR_solar_capacity = DKFR['FR_sol%']

plt.figure(figsize=(10, 6))
plt.plot(day, DK_solar_capacity, label='DK_solar_percentage')
plt.plot(day, FR_solar_capacity, label='FR_solar_percentage')

# This Plot Chart shows the percentage of Solar vs total Capacity in each country of DK & FR.
# The percentage of Solar in FR was started and kept increasing from 2000, and it last in long term, it dropped in 2006, there are other renewable Energy are growing in FR and the total Capacity is growing faster than Solar.
# FR% reahced peak in 2015, and it remained at 33% since then.
# The percentage of Solar in DK slowly started in 2011, reached peak in 2014, it remained at 8% since then.
```

```python
# Plot Chart of DK & FR solar vs all total percentage

day = DKFR['Day']
DK_solar_capacity = DKFR['DK_sol%_total']
FR_solar_capacity = DKFR['FR_sol%_total']

plt.figure(figsize=(10, 6))
plt.plot(day, DK_solar_capacity, label='DK_solar_percentage_all')
plt.plot(day, FR_solar_capacity, label='FR_solar_percentage_all')

# This Plot Chart shows the percentage of Solar of DK and FR vs total Capacity across all countries.
# It shows the similar increasement for DK & FR to the previous chart.
# FR% Solar reached peak in 2012, it Contributed 3.60% of total Cap.
# DK% Solar reached peak in 2014, it Contributed 0.25% of total Cap.
```
